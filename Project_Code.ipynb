```python
#importing tabula
import tabula
#the pd is the standard shorthand for pandas
import pandas as pd
import numpy as np

import PyPDF2
from os import listdir
from os.path import isfile, join
import glob
import pdfplumber
```

```python

final = pd.DataFrame()
for name in glob.glob('E:\Education authority of NI_Internship\BEMs project\Armagh BEMS\**\*.pdf'):
    #try:
        print(name)
        input_analogue= 0
        input_digital= 0
        output_analogue = 0
        output_digital= 0
        manitenance_analogue = 0
        maintenance_digital= 0 
        sch_2=[]
        #s = ','.join(scheduling)
        #lol_string = ''.join(map(str, scheduling))
        with pdfplumber.open(name) as pdf:
            for i in range(len(pdf.pages)):
                tb=pdf.pages[i].extract_text()
                #print(tb)
                x = tb.split()
                #print(x)
                #if 'BEMS'& 'Maintenance' &  'Inputs' in x:
                if 'BEMS' and ' Maintenance' and 'Inputs' in x:
                    i_a=x.count("A") or x.count("Ana")
                    #print(a)
                    input_analogue= input_analogue + i_a
                if 'BEMS' and ' Maintenance' and 'Inputs' in x:
                    i_d=x.count("D") or x.count("Dig")
                    #print(d)
                    input_digital= input_digital + i_d
                if 'BEMS' and ' Maintenance' and 'Outputs' in x:
                    o_a=x.count("A") or x.count("Ana")
                    #print(a)
                    output_analogue= output_analogue + o_a
                if 'BEMS' and ' Maintenance' and 'Outputs' in x:    
                    o_d=x.count("D") or x.count("Dig")
                    #print(d)
                    output_digital= output_digital + o_d
                if 'BEMS' and ' Maintenance' and 'Metering' in x:  
                    m_a=x.count("A") or x.count("Ana")
                    #print(a)
                    manitenance_analogue=manitenance_analogue + m_a
                if 'BEMS' and ' Maintenance' and 'Metering' in x:   
                    m_d=x.count("D") or x.count("Dig")
                    #print(d)
                    maintenance_digital= maintenance_digital + m_d
                if 'BEMS' and 'Maintenance' and 'Time' and 'Schedules' and "Monday" in x:
                    #print(x)
                    sch=[]  
                    target = "Wednesday"
                    start= 2
                    for i,w in enumerate(x):   
                        while w == target:
                            # next word
                            word= x[i+start]
                            sch.append(word)
                            #print(word)
                            i= i + 1
                            if word == "Thursday":
                                break
                        sch_1= " ".join(sch[:-1])
                    #print(sch_1)    
                    sch_2.append(sch_1)
              
        #print(sch_2)       
                #print(analogue)
                #print(digital)
              #  print("------------------------------------PAGE END-----------------------------------------------------")
                #print(name)
            df = tabula.read_pdf(name,pages= "all",)
            lastpage=len(df)-1
            df_last=df[lastpage] 
            #print(df_last)
            #df_last= df_last.dropna(axis=1,how='all')
            #df_last=df_last.reset_index(axis=0)
            df_last_list= df_last.to_dict(orient='list')
           #newDict = dict()
            fault_count = 0
            for key, value in df_last_list.items():
                if "Fault" in value:
                    fault_idx = value.index("Fault")
                    for data in value[fault_idx+1:]:
                        if str(data) != "nan":
                            fault_count += 1
            #print(fault_count)
            comment_count = 0 
            for key, value in df_last_list.items():
                if "Comment" in value:
                    comment_idx = value.index("Comment")
                    for data in value[comment_idx+1:]:
                        if str(data) != "nan":
                            comment_count += 1
            #print(comment_count) 
            recommendation_count = 0 
            for key, value in df_last_list.items():
                if "Recommendations" in value:
                    recommendation_idx = value.index("Recommendations")
                    for data in value[recommendation_idx+1:]:
                        if str(data) != "nan":
                            recommendation_count += 1
            #print(recommendation_count)  
            fault= max(fault_count, comment_count,recommendation_count)  
            #print("max: ", fault) 
            dF_2=pd.DataFrame(df[0])
            final_1=pd.DataFrame({
                          "File Name" : name,
                          "Plant":[dF_2['Unnamed: 0'][0]],
                          "Site":[dF_2['BEMS Maintenance Checklist'][0]],
                          "Controller": [dF_2['Unnamed: 0'][3]],
                          "Date":[dF_2['BEMS Maintenance Checklist'][3]],
                          "Back Up Saved" :[dF_2['Unnamed: 2'][5]],
                          "Network/Communication Device" :[dF_2['Unnamed: 2'][6]],
                          "ODA Frost" :[dF_2['Unnamed: 2'][7]],
                          "Immersion Frost" : [dF_2['Unnamed: 2'][8]],
                          "Room Frost/3rd Stage Frost" :[dF_2 ['Unnamed: 2'][9]],
                          "BST/GMT Start +1HR" : [dF_2['Unnamed: 2'][10]],
                          "Heat Disapation" : [dF_2['Unnamed: 2'][11]],
                          "Panel Lights" : [dF_2['Unnamed: 2'][12]],
                          "Panel Condition" :[dF_2['Unnamed: 2'][13]],
                          "Boiler Lead/Lag" :[dF_2['Unnamed: 2'][14]],
                          "Input Analogue" :  input_analogue,
                          "Input Digital" : input_digital,
                          "Output Analogue" : output_analogue,
                          "Output Digital"  : output_digital,
                          "Maintenance Analogue": manitenance_analogue,
                          "Maintenance Digital" : maintenance_digital,
                          "Scheduling" : ','.join(sch_2),
                          "Fault" : fault
                         })
            #print(scheduling)
            print("-------------------------------------------FILE END-------------------------------------------------------")
        final=final.append(final_1,ignore_index=True)
        #print(name)
    #except:
     #   print("----------------", name ,"------------------------------------")
      #  print("Error")
   
     
```
